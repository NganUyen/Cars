## Clean engine_type

```python
def detect_engine_layout(engine):
    engine = str(engine).upper()

   #clarify elec and hybird
    if 'ELECTRIC' in engine:
        return 'Electric Motor'
    if 'HYBRID' in engine:
        return 'Hybrid (ICE+Electric)'

    #clarify V, I, W, Boxer of car:
    if 'V' in engine and any(ch.isdigit() for ch in engine):
        return 'V-type'
    if any(word in engine for word in ['I3', 'I4', 'I5', 'I6', 'INLINE', 'CYLINDER']):
        return 'Inline'
    if 'FLAT' in engine or 'BOXER' in engine:
        return 'Boxer'
    if 'W' in engine and any(ch.isdigit() for ch in engine):
        return 'W-type'

    # 3. Some another 
    if any(word in engine for word in ['MIVEC', 'SKYACTIV', 'REVOTRON']):
        return 'Inline'

    # 4. fallback
    return 'Inline'
```

```python
data["Engine_Layout"] = data["Engine_Type"].apply(detect_engine_layout)
```

## Clean Fuel_Type

```python
def normalize_fuel(fuel):
    f = str(fuel).lower().strip()
    if "electric" in f and "hybrid" not in f:
        return "Electric"
    elif "hybrid" in f:
        return "Hybrid"
    elif "diesel" in f and "petrol" in f:
        return "Mixed Fuel"
    elif "diesel" in f:
        return "Diesel"
    elif "petrol" in f or "gasoline" in f:
        return "Petrol"
    elif "hydrogen" in f or "cng" in f:
        return "Hydrogen/CNG"
    else:
        return "Other"

data["Fuel_Type_Cleaned"] = data["Fuel_Type"].apply(normalize_fuel)
data["Fuel_Type_Cleaned"].value_counts()
```
